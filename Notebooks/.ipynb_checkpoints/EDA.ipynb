# 📊 Bank Liquidity Anomaly Detection - EDA

This notebook performs Exploratory Data Analysis (EDA) on banking transaction data to understand patterns in liquidity inflow and outflow.

The goal is to:
- Understand transaction behavior  
- Identify temporal patterns  
- Justify feature engineering choices  
- Prepare for anomaly detection

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(
    "../../DATA/RAW/financial_fraud_detection_dataset(1).csv",
    usecols=['timestamp','amount','transaction_type'],
    nrows=300000
)
```

```python
df.shape
df.info()
df.describe()
```

### Insights:
- Dataset contains transaction-level data
- Includes timestamp, amount, and transaction type
- No severe missing values observed

```python
df.isnull().sum()
```

### Insights:
- Missing values are minimal and handled during preprocessing

```python
df['transaction_type'].value_counts().plot(kind='bar')
plt.title("Transaction Type Distribution")
plt.show()
```

### Insights:
- Deposits and withdrawals are unevenly distributed
- Helps in understanding liquidity behavior

```python
df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')
df = df.dropna(subset=['timestamp'])

df = df.sort_values('timestamp')
```

```python
df['deposit_amount'] = np.where(df['transaction_type']=='deposit', df['amount'], 0)
df['withdrawal_amount'] = np.where(df['transaction_type']!='deposit', df['amount'], 0)
```

```python
df.set_index('timestamp', inplace=True)

df_hourly = df.resample('h').agg({
    'deposit_amount':'sum',
    'withdrawal_amount':'sum'
})

df_hourly['net_flow'] = df_hourly['deposit_amount'] - df_hourly['withdrawal_amount']

df_hourly.head()
```

```python
plt.figure(figsize=(15,5))
plt.plot(df_hourly['net_flow'])
plt.title("Net Liquidity Flow Over Time")
plt.show()
```

### Insights:
- Net flow shows high fluctuations
- Sudden spikes may indicate anomalies
- Strong time-based patterns exist

```python
df_hourly['hour'] = df_hourly.index.hour

df_hourly.groupby('hour')['net_flow'].mean().plot()
plt.title("Average Net Flow by Hour")
plt.show()
```

### Insights:
- Transactions vary significantly by hour
- Indicates strong cyclical behavior

```python
df_hourly['day_of_week'] = df_hourly.index.dayofweek
df_hourly['is_weekend'] = df_hourly['day_of_week'].isin([5,6]).astype(int)

df_hourly.groupby('is_weekend')['net_flow'].mean().plot(kind='bar')
plt.title("Weekend vs Weekday Net Flow")
plt.show()
```

### Insights:
- Weekend behavior differs from weekdays
- Useful feature for model training

```python
df_hourly['net_flow'].hist(bins=50)
plt.title("Distribution of Net Flow")
plt.show()
```

### Insights:
- Distribution is highly skewed
- Presence of extreme values (potential anomalies)

```python
df_hourly['rolling_mean_24'] = df_hourly['net_flow'].rolling(24).mean()

df_hourly['rolling_mean_24'].plot()
plt.title("Rolling Mean (24 hours)")
plt.show()
```

### Insights:
- Smooth trend observed over time
- Justifies use of rolling features

# 📌 Final Insights

- Transaction behavior is highly time-dependent  
- Strong hourly and weekly patterns observed  
- Net flow distribution is skewed with extreme values  
- Rolling trends indicate temporal dependencies  

### 💡 Conclusion:
These insights justify the use of:
- Time-based features (hour, weekday)
- Lag features (lag_1, lag_24)
- Rolling statistics (mean, std)
- Anomaly detection models (Isolation Forest)
