```python
import os
import pandas as pd

# Check current path
print("Current Path:", os.getcwd())

# Go TWO levels up
BASE_DIR = os.path.dirname(os.path.dirname(os.getcwd()))

file_path = os.path.join(BASE_DIR, "DATA", "Processed_data", "df_final.csv")

print("Final Path:", file_path)

df = pd.read_csv(file_path)

# Fix date column
if 'Unnamed: 0' in df.columns:
    df.rename(columns={'Unnamed: 0': 'date'}, inplace=True)

df['date'] = pd.to_datetime(df['date'], errors='coerce')
df.set_index('date', inplace=True)

df.head()
```

```python
import numpy as np

numeric_df = df.select_dtypes(include=[np.number])
numeric_df.columns
```

```python
corr_matrix = numeric_df.corr()
corr_matrix
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,7
                    ))

sns.heatmap(
    corr_matrix,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    linewidths=0.5
)

plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
corr_matrix = numeric_df.corr()

corr_with_netflow = corr_matrix['net_flow'].sort_values(ascending=False)

corr_with_netflow
```

```python
corr_with_netflow.drop('net_flow').plot(kind='bar', figsize=(10,5))

plt.title("Correlation with Net Flow")
plt.ylabel("Correlation")
plt.show()
```

## 📊 Correlation Analysis of Features

To understand the relationships between different variables in the dataset, a correlation analysis was performed with respect to the target variable **net_flow**. The correlation values indicate how strongly each feature is linearly related to net flow.

---

### 🔍 Key Observations

#### ✅ Strong Positive Correlation
- **deposit_amount (~0.9+)** shows the highest positive correlation with net flow, indicating that deposits are the primary contributor to liquidity.
- **amount (~0.75–0.8)** also has a strong positive relationship, suggesting that transaction volume significantly impacts net flow.
- **net_flow_diff (~0.68)** and **pct_change (~0.64)** capture recent changes and trends, making them highly informative features.

---

#### 🟡 Moderate Correlation
- **rolling_mean_24 (~0.27)** represents the moving average trend and has a moderate influence on net flow.
- **inflation_rate (~0.13)** shows a weak but positive relationship, indicating limited short-term macroeconomic impact.

---

#### 🔵 Weak or Near-Zero Correlation
- Features such as **lag_1**, **lag_24**, **hour_sin**, **hour_cos**, and **interest_rate** exhibit very low correlation with net flow.
- This suggests that time-based patterns and some historical values may not have a strong direct linear influence.

---

#### 🔴 Negative Correlation
- **withdrawal_amount (~ -0.18)** has a negative correlation, indicating that higher withdrawals reduce net liquidity.
- **rolling_std_24** shows slight negative correlation, suggesting that higher volatility may reduce stability.

---

### 🧠 Insights

- Liquidity is primarily driven by **deposit and transaction-related features**.
- **Derived features** such as percentage change and differences play a crucial role in capturing dynamic behavior.
- **Time-based and macroeconomic features** have minimal direct impact on short-term liquidity.
- Some features with low correlation are still retained, as anomaly detection models like Isolation Forest can capture **non-linear relationships**.

---

### 🎯 Conclusion

The correlation analysis helped in identifying the most influential features and understanding their relationships with net flow. This ensured better feature selection and contributed to building a more robust anomaly detection system.
