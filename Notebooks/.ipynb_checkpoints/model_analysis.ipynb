```python
import joblib

model = joblib.load("../../models/isolation_forest.pkl")
scaler = joblib.load("../../models/scaler.pkl")
```

```python
import pandas as pd

df_hourly = pd.read_csv("../../DATA/Processed_data/df_final.csv")

df_hourly.rename(columns={'Unnamed: 0': 'date'}, inplace=True)

df_hourly['date'] = pd.to_datetime(df_hourly['date'])

df_hourly.set_index('date', inplace=True)
```

```python
df_hourly.head()
```

```python
scaler.feature_names_in_
```

```python
features = list(scaler.feature_names_in_)
```

```python
X = df_hourly[features].dropna()
X_scaled = scaler.transform(X)
```

```python
df_anomaly = X.copy()

df_anomaly['anomaly'] = model.predict(X_scaled)
df_anomaly['anomaly'] = df_anomaly['anomaly'].map({1:0, -1:1})
```

```python
df_anomaly['net_flow'] = df_hourly.loc[df_anomaly.index, 'net_flow']
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,5))

plt.plot(df_anomaly['net_flow'], label='Normal')
plt.scatter(
    df_anomaly[df_anomaly['anomaly']==1].index,
    df_anomaly[df_anomaly['anomaly']==1]['net_flow'],
    color='red',
    label='Anomaly'
)

plt.title("Anomaly Detection in Net Flow")
plt.legend()
plt.show()
```

```python
df_anomaly['anomaly'].value_counts(normalize=True)
```

### Insights:
- Only a small percentage of data points are classified as anomalies
- This aligns with real-world expectations where anomalies are rare

```python
import seaborn as sns
sns.boxplot(x=df_anomaly['anomaly'], y=df_anomaly['net_flow'])
plt.title("Normal vs Anomaly Distribution")
plt.show()
```

### Insights:
- Anomalies show extreme values compared to normal data
- Confirms model is capturing unusual liquidity behavior

```python
df_anomaly['hour'] = df_anomaly.index.hour

sns.countplot(x='hour', hue='anomaly', data=df_anomaly)
plt.title("Anomalies by Hour")
plt.show()
```

### Insights:
- Anomalies are more frequent at certain hours
- Indicates time-based irregular patterns

# 📌 Model-Based Insights

- Isolation Forest successfully identified unusual spikes in liquidity
- Anomalies correspond to extreme net flow values
- Temporal patterns suggest anomalies occur at specific times
- Model captures both magnitude and sudden changes

### 💡 Conclusion:
The anomaly detection system effectively identifies abnormal liquidity behavior, making it useful for fraud detection and financial monitoring.
